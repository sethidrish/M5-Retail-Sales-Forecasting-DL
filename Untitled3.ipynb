```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# --- Global Parameters ---
LOOK_BACK_WINDOW = 28 # Number of past timesteps to consider for prediction
# print(f"Global parameter LOOK_BACK_WINDOW set to: {LOOK_BACK_WINDOW}") # Removed conversational print

# --- Mount Google Drive ---
from google.colab import drive
drive.mount('/content/drive')
# print("Google Drive mounted.") # Removed conversational print

# --- Define the path to your data folder ---
# This path points to the 'data' folder directly inside your 'My Drive' in Google Drive.
DATA_PATH = '/content/drive/MyDrive/data/'
# print(f"DATA_PATH set to: {DATA_PATH}") # Removed conversational print


# print("\n--- Starting Data Loading and Early Filtering ---") # Removed conversational print

# 1. Load sales_train_validation.csv
sales_df = pd.read_csv(os.path.join(DATA_PATH, 'sales_train_validation.csv'))

# 2. Load calendar.csv
calendar_df = pd.read_csv(os.path.join(DATA_PATH, 'calendar.csv'))

# 3. Load sell_prices.csv
sell_prices_df = pd.read_csv(os.path.join(DATA_PATH, 'sell_prices.csv'))

# print("Initial datasets loaded. Performing early filtering...") # Removed conversational print

# --- CRITICAL: EXTREME DATASET REDUCTION (EARLY APPLICATION ON sales_df) ---
# This is done on sales_df (wide format) before melting to keep memory low from the start.
selected_item_ids = [
    'HOBBIES_1_001', 'HOBBIES_1_002', 'HOBBIES_1_003',
    'HOBBIES_1_004', 'HOBBIES_1_005', 'HOBBIES_1_006',
    'HOBBIES_1_007', 'HOBBIES_1_008', 'HOBBIES_1_009',
    'HOBBIES_1_010'
]
selected_store = 'CA_1'

initial_rows_sales_df = sales_df.shape[0]
sales_df = sales_df[
    (sales_df['store_id'] == selected_store) &
    (sales_df['item_id'].isin(selected_item_ids))
].reset_index(drop=True)

# print(f"Sales data reduced from {initial_rows_sales_df:,} rows to {sales_df.shape[0]:,} rows BEFORE melting.") # Removed conversational print
# print(f"Working with store: '{selected_store}' and specific items: {selected_item_ids[0]}...{selected_item_ids[-1]}.") # Removed conversational print


# --- Reshape sales_df (now tiny) from Wide to Long Format ---
id_columns_filtered = ['id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id']
sales_columns_filtered = [col for col in sales_df.columns if col.startswith('d_')]

sales_melt_df = sales_df.melt(id_vars=id_columns_filtered,
                              value_vars=sales_columns_filtered,
                              var_name='d',
                              value_name='sales')
# print(f"sales_melt_df created with shape: {sales_melt_df.shape}") # Removed conversational print

# Convert 'date' column in calendar_df to datetime objects.
calendar_df['date'] = pd.to_datetime(calendar_df['date'])

# --- Merge DataFrames (now on tiny sales_melt_df) ---
full_df = pd.merge(sales_melt_df, calendar_df, on='d', how='left')
full_df = pd.merge(full_df, sell_prices_df, on=['store_id', 'item_id', 'wm_yr_wk'], how='left')
# print(f"full_df (after all merges) created with shape: {full_df.shape}") # Removed conversational print


# --- Encode Categorical ID columns to numerical codes for Embedding layers. ---
id_cols_to_encode = ['item_id', 'dept_id', 'cat_id', 'store_id', 'state_id']
for col in id_cols_to_encode:
    if col in full_df.columns:
        full_df[f'{col}_encoded'] = full_df[col].astype('category').cat.codes
    # else:
        # print(f"DEBUG Warning: Original ID column '{col}' not found for encoding.") # Removed debug print
# print("Categorical ID columns encoded.") # Removed conversational print


# --- Handle Missing Sell Prices & Optimize Data Types (on tiny full_df) ---
full_df = full_df.sort_values(by=['item_id', 'store_id', 'date']).reset_index(drop=True)
full_df['sell_price'] = full_df.groupby(['item_id', 'store_id'])['sell_price'].transform(lambda x: x.ffill().bfill())
full_df['sell_price'] = full_df['sell_price'].fillna(0) # Fill any remaining NaNs with 0.

# Function to downcast numeric types and convert objects to categories for memory efficiency.
def reduce_mem_usage(df):
    # start_mem = df.memory_usage().sum() / 1024**2 # Removed debug print
    for col in df.columns:
        col_type = df[col].dtype
        if col_type != object and col_type.name != 'category' and 'datetime' not in col_type.name:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)
            else: # float
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        elif 'datetime' not in col_type.name:
            df[col] = df[col].astype('category')
    # end_mem = df.memory_usage().sum() / 1024**2 # Removed debug print
    # print(f"DEBUG: reduce_mem_usage - Start Memory: {start_mem:.2f} MB") # Removed debug print
    # print(f"DEBUG: reduce_mem_usage - End Memory: {end_mem:.2f} MB") # Removed debug print
    # print(f"DEBUG: reduce_mem_usage - Decreased by {100 * (start_mem - end_mem) / start_mem:.2f}%") # Removed debug print
    return df

full_df = reduce_mem_usage(full_df)

# --- END OF CELL 1 ---
```

```python
import numpy as np # Ensure numpy is imported for later operations like np.sort, np.int32 etc.

# --- Extract Time-Based Features ---
# Ensure 'date' is a datetime type (it should be from Cell 1).
full_df['date'] = pd.to_datetime(full_df['date'])

# Extract various time components as new numerical features.
full_df['day'] = full_df['date'].dt.day.astype(np.int8)
full_df['dayofweek'] = full_df['date'].dt.dayofweek.astype(np.int8) # Monday=0, Sunday=6
full_df['dayofyear'] = full_df['date'].dt.dayofyear.astype(np.int16)
full_df['weekofyear'] = full_df['date'].dt.isocalendar().week.astype(np.int8) # Using isocalendar for week number
full_df['quarter'] = full_df['date'].dt.quarter.astype(np.int8)
full_df['is_month_start'] = full_df['date'].dt.is_month_start.astype(np.int8) # Convert boolean to int
full_df['is_month_end'] = full_df['date'].dt.is_month_end.astype(np.int8) # Convert boolean to int

# --- Create Lag Features for Sales ---
# Define the lag periods.
LAG_FEATURES = [7, 28, 56]

# Sort the DataFrame to ensure correct lag calculation within each item-store group.
full_df = full_df.sort_values(by=['id', 'date']).reset_index(drop=True)

for lag in LAG_FEATURES:
    full_df[f'sales_lag_{lag}'] = full_df.groupby('id')['sales'].shift(lag).astype(np.float16)

# --- Create Rolling Window Features for Sales ---
# Define the rolling window periods.
ROLLING_WINDOWS = [7, 28]

for window in ROLLING_WINDOWS:
    full_df[f'rolling_mean_sales_{window}'] = full_df.groupby('id')['sales'].transform(lambda x: x.rolling(window=window).mean().astype(np.float16))
    full_df[f'rolling_std_sales_{window}'] = full_df.groupby('id')['sales'].transform(lambda x: x.rolling(window=window).std().astype(np.float16))

# --- Fill Remaining NaN Values in New Features ---
# Identify columns that were just created and contain NaNs (lags and rolling features).
new_feature_columns = [col for col in full_df.columns if 'sales_lag_' in col or 'rolling_mean_sales_' in col or 'rolling_std_sales_' in col]
for col in new_feature_columns:
    full_df[col] = full_df[col].fillna(0) # Fill all NaNs in these new features with 0.

# --- One-Hot Encode Lower Cardinality Categorical Features ---
# Define the columns to one-hot encode.
categorical_features_to_encode = [
    'weekday', 'wday', 'month', 'year', 'quarter',
    'event_name_1', 'event_type_1', 'event_name_2', 'event_type_2',
    'snap_CA', 'snap_TX', 'snap_WI'
]
# Convert to 'category' dtype for efficiency before one-hot encoding.
for col in categorical_features_to_encode:
    if col in full_df.columns:
        full_df[col] = full_df[col].astype('category')

# Perform one-hot encoding.
full_df = pd.get_dummies(full_df, columns=categorical_features_to_encode, dummy_na=False)

# --- END OF CELL 2 ---
```

```python
import numpy as np # Ensure numpy is imported if not already.
from sklearn.preprocessing import MinMaxScaler # Ensure MinMaxScaler is available.

# --- Data Splitting (Final Refined Logic) ---
# Sort the data by date and then by item ID to ensure proper chronological order
full_df = full_df.sort_values(by=['date', 'id']).reset_index(drop=True)

# Define our target variable
TARGET = 'sales'

# Define the features (X) for our model.
# Use the _encoded ID columns and exclude original string IDs and 'id' itself for sequential features.
# CRITICAL FIX: Include 'date' in features_for_X_df so it's carried through for plotting.
features = [col for col in full_df.columns if col not in ['id', 'item_id_encoded', 'dept_id_encoded', 'cat_id_encoded', 'store_id_encoded', 'state_id_encoded', 'd', 'date', 'sales', 'wm_yr_wk']]
encoded_id_features = ['item_id_encoded', 'dept_id_encoded', 'cat_id_encoded', 'store_id_encoded', 'state_id_encoded']

# Combine original features with the encoded ID features and 'id' AND 'date' for X_train etc.
features_for_X_df = features + encoded_id_features + ['id', 'date'] # Added 'date' here

# Get all unique dates in the current (possibly reduced) full_df
unique_dates = full_df['date'].unique()
unique_dates = np.sort(unique_dates)
unique_dates = pd.to_datetime(unique_dates)

total_unique_days_in_df = len(unique_dates)

# Define proportions for splits
TRAIN_RATIO = 0.8
VAL_RATIO = 0.1
TEST_RATIO = 0.1

# Calculate initial split indices based on proportions
train_split_idx = int(total_unique_days_in_df * TRAIN_RATIO)
val_split_idx = int(total_unique_days_in_df * (TRAIN_RATIO + VAL_RATIO))

# Ensure minimum number of days for training after accounting for look-back window
min_required_days_for_train = LOOK_BACK_WINDOW + 1

if train_split_idx < min_required_days_for_train:
    train_split_idx = min_required_days_for_train

if val_split_idx <= train_split_idx:
    val_split_idx = train_split_idx + max(1, int(total_unique_days_in_df * VAL_RATIO / 2))
    if val_split_idx >= total_unique_days_in_df:
        val_split_idx = total_unique_days_in_df - max(1, int(total_unique_days_in_df * TEST_RATIO / 2))
        if val_split_idx <= train_split_idx:
            train_split_idx = val_split_idx - 1
            if train_split_idx < 0:
                train_split_idx = 0

# Convert indices to actual dates
train_split_date = unique_dates[train_split_idx]
val_split_date = unique_dates[val_split_idx]

# Perform the splits using the calculated dates
X_train = full_df[full_df['date'] < train_split_date][features_for_X_df]
y_train = full_df[full_df['date'] < train_split_date][TARGET]

X_val = full_df[(full_df['date'] >= train_split_date) & (full_df['date'] < val_split_date)][features_for_X_df]
y_val = full_df[(full_df['date'] >= train_split_date) & (full_df['date'] < val_split_date)][TARGET]

X_test = full_df[full_df['date'] >= val_split_date][features_for_X_df]
y_test = full_df[full_df['date'] >= val_split_date][TARGET]


# --- Feature Scaling ---
# Identify numerical columns for scaling.
numerical_cols_to_scale = [
    'sell_price', 'day', 'dayofweek', 'dayofyear', 'weekofyear', 'quarter',
    'is_month_start', 'is_month_end',
    'sales_lag_7', 'sales_lag_28', 'sales_lag_56',
    'rolling_mean_sales_7', 'rolling_std_sales_7',
    'rolling_mean_sales_28', 'rolling_std_sales_28'
]
# CRITICAL FIX for sequential_features below: Exclude original string IDs AND 'date'
# This list is used for scaler.fit/transform as well.
numerical_cols_to_scale = [col for col in numerical_cols_to_scale if col in X_train.columns]

# Initialize and fit scaler ONLY on X_train.
scaler = MinMaxScaler()
scaler.fit(X_train[numerical_cols_to_scale])

# Transform all datasets.
X_train_scaled = X_train.copy()
X_val_scaled = X_val.copy()
X_test_scaled = X_test.copy()
X_train_scaled[numerical_cols_to_scale] = scaler.transform(X_train[numerical_cols_to_scale])
X_val_scaled[numerical_cols_to_scale] = scaler.transform(X_val[numerical_cols_to_scale])
X_test_scaled[numerical_cols_to_scale] = scaler.transform(X_test[numerical_cols_to_scale])


# --- Creating Sequences for Deep Learning Input ---
# Define the columns used as sequential features (excluding IDs and 'date').
sequential_features = [col for col in X_train_scaled.columns if col not in [
    'id', 'item_id', 'dept_id', 'cat_id', 'store_id', 'state_id', # Original string IDs
    'item_id_encoded', 'dept_id_encoded', 'cat_id_encoded', 'store_id_encoded', 'state_id_encoded', # Encoded IDs
    'date' # Exclude 'date' from sequential features as it's not a numeric feature for LSTM input
]]

# Define the *encoded* categorical ID columns that need to be returned for embedding inputs.
categorical_id_columns_for_embedding = ['item_id_encoded', 'dept_id_encoded', 'cat_id_encoded', 'store_id_encoded', 'state_id_encoded']

# Function to create sequences for a given dataframe, including encoded ID features and TARGET DATES for embedding/plotting.
def create_sequences_with_ids(data_df, target_series, look_back):
    X, y, dates_for_seq = [], [], [] # Added dates_for_seq
    item_ids_enc, dept_ids_enc, cat_ids_enc, store_ids_enc, state_ids_enc = [], [], [], [], []

    for item_store_id_val in data_df['id'].unique():
        item_series_full = data_df[data_df['id'] == item_store_id_val]

        item_series_features_only = item_series_full[sequential_features]
        item_target = target_series[data_df['id'] == item_store_id_val]

        item_series_np = item_series_features_only.to_numpy()
        item_target_np = item_target.to_numpy()

        # We need the index of item_series_full to use .at correctly
        item_series_full_index = item_series_full.index.to_numpy() # Get the underlying numpy array of the index values

        for i in range(len(item_series_np) - look_back):
            feature_sequence = item_series_np[i:(i + look_back)]
            target_value = item_target_np[i + look_back]

            target_row_global_idx = item_series_full_index[i + look_back] # Get the global index of the target row

            # Extract scalar integer directly from DataFrame using .at
            item_id_val_for_seq = full_df.at[target_row_global_idx, 'item_id_encoded']
            dept_id_val_for_seq = full_df.at[target_row_global_idx, 'dept_id_encoded']
            cat_id_val_for_seq = full_df.at[target_row_global_idx, 'cat_id_encoded']
            store_id_val_for_seq = full_df.at[target_row_global_idx, 'store_id_encoded']
            state_id_val_for_seq = full_df.at[target_row_global_idx, 'state_id_encoded']

            # Added: Extract date for the target value for plotting
            date_for_seq = full_df.at[target_row_global_idx, 'date'] # Get the date from full_df

            X.append(feature_sequence)
            y.append(target_value)
            dates_for_seq.append(date_for_seq) # Append date for this sequence's target

            item_ids_enc.append(item_id_val_for_seq)
            dept_ids_enc.append(dept_id_val_for_seq)
            cat_ids_enc.append(cat_id_val_for_seq)
            store_ids_enc.append(store_id_val_for_seq)
            state_ids_enc.append(state_id_val_for_seq)

    # Convert lists to numpy arrays and reshape for Keras input (samples, 1 or 3D)
    num_samples = len(X) # This is the correct number of sequences generated

    return np.array(X, dtype=np.float32), np.array(y, dtype=np.float32), \
           np.array(item_ids_enc, dtype=np.int32).reshape(num_samples, 1), \
           np.array(dept_ids_enc, dtype=np.int32).reshape(num_samples, 1), \
           np.array(cat_ids_enc, dtype=np.int32).reshape(num_samples, 1), \
           np.array(store_ids_enc, dtype=np.int32).reshape(num_samples, 1), \
           np.array(state_ids_enc, dtype=np.int32).reshape(num_samples, 1), \
           np.array(dates_for_seq) # Return dates as a NumPy array


# Data generation for training, validation, and test sets.
X_train_sequences, y_train_sequences, \
item_id_train_seq, dept_id_train_seq, cat_id_train_seq, store_id_train_seq, state_id_train_seq, \
dates_train_seq = \
create_sequences_with_ids(X_train_scaled, y_train, LOOK_BACK_WINDOW)

X_val_sequences, y_val_sequences, \
item_id_val_seq, dept_id_val_seq, cat_id_val_seq, store_id_val_seq, state_id_val_seq, \
dates_val_seq = \
create_sequences_with_ids(X_val_scaled, y_val, LOOK_BACK_WINDOW)

X_test_sequences, y_test_sequences, \
item_id_test_seq, dept_id_test_seq, cat_id_test_seq, store_id_test_seq, state_id_test_seq, \
dates_test_seq = \
create_sequences_with_ids(X_test_scaled, y_test, LOOK_BACK_WINDOW)


# Print shapes of generated data.
print("\nShapes of generated data:")
print(f"  X_train_sequences: {X_train_sequences.shape}, y_train_sequences: {y_train_sequences.shape}")
print(f"  item_id_train_seq: {item_id_train_seq.shape}, dept_id_train_seq: {dept_id_train_seq.shape}, cat_id_train_seq: {cat_id_train_seq.shape}, store_id_train_seq: {store_id_train_seq.shape}, state_id_train_seq: {state_id_train_seq.shape}")
print(f"  dates_train_seq: {dates_train_seq.shape}")

print(f"  X_val_sequences: {X_val_sequences.shape}, y_val_sequences: {y_val_sequences.shape}")
print(f"  item_id_val_seq: {item_id_val_seq.shape}, dept_id_val_seq: {dept_id_val_seq.shape}, cat_id_val_seq: {cat_id_val_seq.shape}, store_id_val_seq: {store_id_val_seq.shape}, state_id_val_seq: {state_id_val_seq.shape}")
print(f"  dates_val_seq: {dates_val_seq.shape}")

print(f"  X_test_sequences: {X_test_sequences.shape}, y_test_sequences: {y_test_sequences.shape}")
print(f"  item_id_test_seq: {item_id_test_seq.shape}, dept_id_test_seq: {dept_id_test_seq.shape}, cat_id_test_seq: {cat_id_test_seq.shape}, store_id_test_seq: {store_id_test_seq.shape}, state_id_test_seq: {state_id_test_seq.shape}")
print(f"  dates_test_seq: {dates_test_seq.shape}")

# --- END OF CELL 3 ---
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Concatenate, Embedding, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# --- Design and Compile the Deep Learning Model ---

# Identify High-Cardinality Categorical Features for Embedding.
original_id_features_for_vocab = ['item_id', 'dept_id', 'cat_id', 'store_id', 'state_id']

# Get the number of unique values for each original categorical ID feature from the (filtered) full_df.
item_id_vocab_size = full_df['item_id'].nunique() + 1
dept_id_vocab_size = full_df['dept_id'].nunique() + 1
cat_id_vocab_size = full_df['cat_id'].nunique() + 1
store_id_vocab_size = full_df['store_id'].nunique() + 1
state_id_vocab_size = full_df['state_id'].nunique() + 1

# Define embedding dimensions (common rule of thumb: min(50, number_of_categories // 2), ensure at least 1).
item_embedding_dim = max(1, min(50, item_id_vocab_size // 2))
dept_embedding_dim = max(1, min(50, dept_id_vocab_size // 2))
cat_embedding_dim = max(1, min(50, cat_id_vocab_size // 2))
store_embedding_dim = max(1, min(50, store_id_vocab_size // 2))
state_embedding_dim = max(1, min(50, state_id_vocab_size // 2))

print(f"Calculated Vocab Sizes: item={item_id_vocab_size}, dept={dept_id_vocab_size}, cat={cat_id_vocab_size}, store={store_id_vocab_size}, state={state_id_vocab_size}")
print(f"Embedding dimensions: item={item_embedding_dim}, dept={dept_embedding_dim}, cat={cat_embedding_dim}, store={store_embedding_dim}, state={state_embedding_dim}")


# Define the Model Architecture.

# Input for sequential numerical/one-hot features.
sequential_input_shape = (LOOK_BACK_WINDOW, X_train_sequences.shape[2])
sequential_input = Input(shape=sequential_input_shape, name='sequential_features_input')

# Inputs for *encoded* categorical ID features.
item_id_input = Input(shape=(1,), name='item_id_input')
dept_id_input = Input(shape=(1,), name='dept_id_input')
cat_id_input = Input(shape=(1,), name='cat_id_input')
store_id_input = Input(shape=(1,), name='store_id_input')
state_id_input = Input(shape=(1,), name='state_id_input')

# Embedding layers for categorical IDs.
item_id_embedding = Embedding(input_dim=item_id_vocab_size, output_dim=item_embedding_dim, name='item_embedding')(item_id_input)
dept_id_embedding = Embedding(input_dim=dept_id_vocab_size, output_dim=dept_embedding_dim, name='dept_embedding')(dept_id_input)
cat_id_embedding = Embedding(input_dim=cat_id_vocab_size, output_dim=cat_embedding_dim, name='cat_embedding')(cat_id_input)
store_id_embedding = Embedding(input_dim=store_id_vocab_size, output_dim=store_embedding_dim, name='store_embedding')(store_id_input)
state_id_embedding = Embedding(input_dim=state_id_vocab_size, output_dim=state_embedding_dim, name='state_embedding')(state_id_input)

# Flatten embedding outputs for concatenation with LSTM output.
item_id_flat = Flatten(name='item_embedding_flat')(item_id_embedding)
dept_id_flat = Flatten(name='dept_embedding_flat')(dept_id_embedding)
cat_id_flat = Flatten(name='cat_embedding_flat')(cat_id_embedding)
store_id_flat = Flatten(name='store_embedding_flat')(store_id_embedding)
state_id_flat = Flatten(name='state_embedding_flat')(state_id_embedding)


# LSTM layer for sequential features.
lstm_out = LSTM(units=128, return_sequences=False, name='lstm_layer')(sequential_input)

# Concatenate LSTM output with flattened embedding outputs.
concatenated_features = Concatenate(name='concat_all_features')([
    lstm_out,
    item_id_flat,
    dept_id_flat,
    cat_id_flat,
    store_id_flat,
    state_id_flat
])

# Dense layers for final prediction.
dense_layer_1 = Dense(units=64, activation='relu', name='dense_1')(concatenated_features)
output_layer = Dense(units=1, activation='relu', name='output_layer')(dense_layer_1) # Output is sales count (non-negative)

# Create the model.
model = Model(inputs=[sequential_input, item_id_input, dept_id_input, cat_id_input, store_id_input, state_id_input],
              outputs=output_layer,
              name='M5_Forecasting_Model')

# Compile the Model.
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

print("\nModel architecture summary:")
model.summary()

# --- END OF CELL 4 ---
```

```python
# --- Prepare Data for Multi-Input Model Training ---

# Define the inputs dictionary for model.fit()
# The keys must match the `name` arguments of the Input layers in the model (defined in Cell 4).
# The values are the NumPy arrays generated in Cell 3.
X_train_inputs = {
    'sequential_features_input': X_train_sequences,
    'item_id_input': item_id_train_seq,
    'dept_id_input': dept_id_train_seq,
    'cat_id_input': cat_id_train_seq,
    'store_id_input': store_id_train_seq,
    'state_id_input': state_id_train_seq
}

X_val_inputs = {
    'sequential_features_input': X_val_sequences,
    'item_id_input': item_id_val_seq,
    'dept_id_input': dept_id_val_seq,
    'cat_id_input': cat_id_val_seq,
    'store_id_input': store_id_val_seq,
    'state_id_input': state_id_val_seq
}

X_test_inputs = {
    'sequential_features_input': X_test_sequences,
    'item_id_input': item_id_test_seq,
    'dept_id_input': dept_id_test_seq,
    'cat_id_input': cat_id_test_seq,
    'store_id_input': store_id_test_seq,
    'state_id_input': state_id_test_seq
}

print("Data inputs prepared for multi-input Keras model.")
print(f"X_train_inputs keys and shapes: {[ (k, v.shape) for k,v in X_train_inputs.items() ]}")
print(f"y_train_sequences shape: {y_train_sequences.shape}")

# --- END OF CELL 5 ---
```

```python
# --- Train the Deep Learning Model ---

# Define callbacks for efficient training.
# EarlyStopping: Stops training if validation MAE does not improve for a certain number of epochs.
early_stopping = EarlyStopping(monitor='val_mae', patience=10, restore_best_weights=True, verbose=1)

# ReduceLROnPlateau: Reduces the learning rate when the validation MAE stops improving.
reduce_lr = ReduceLROnPlateau(monitor='val_mae', factor=0.5, patience=5, min_lr=0.00001, verbose=1)

# Number of training epochs (iterations over the entire dataset).
EPOCHS = 50

# Batch size: Number of samples per gradient update. Adjust based on GPU memory.
BATCH_SIZE = 256


print(f"Starting model training for {EPOCHS} epochs with batch size {BATCH_SIZE}...")
print(f"Monitoring 'val_mae' for EarlyStopping (patience={early_stopping.patience}) and ReduceLROnPlateau (patience={reduce_lr.patience}).")

# Train the model using the prepared inputs and target.
# Pass a dictionary of inputs for X, as our model has multiple input branches.
history = model.fit(
    x=X_train_inputs,
    y=y_train_sequences,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_data=(X_val_inputs, y_val_sequences), # Provide validation data for monitoring
    callbacks=[early_stopping, reduce_lr], # Apply the defined callbacks
    verbose=1 # Show training progress
)

print("\nModel training complete.")

# --- END OF CELL 6 ---
```
